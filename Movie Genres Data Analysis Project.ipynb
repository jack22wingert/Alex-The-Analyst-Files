```python
# remove duplicate rows
movies.drop_duplicates(inplace = True)
```

```python
#drop rows with null values in the genres column
movies.dropna(subset = ['genres'], inplace = True)
```

```python
# add a profit column to the data
movies['profit'] = movies['revenue'] - movies['budget']
movies.head()
```

```python
#keeping columns relevant to our data 
movies_genre = movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
movies_genre.head()
```

```python
#breaking up the genre column for analysis
from pandas import Series, DataFrame
```

```python
# split up genres into their own row for data analysis
split = movies_genre['genres'].str.split('|').apply(lambda x: Series(x)).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
movies_genre.head()
```

### Research Hypotheses (H):
5. Profit of genre per year.

```python
movies_genre_2 = movies[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]

# split up genres into their own row for data analysis
split = movies_genre_2['genres'].str.split('|').apply(lambda x: Series(x)).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre_2['genres']
movies_genre_2 = movies_genre_2.join(split)
```

```python
movies_genre_2.head()
```

```python
time_genre = pd.DataFrame(movies_genre_2.groupby(['release_year','genres_split'])['profit'].mean())
time_genre.head()
```

```python
final_genre = pd.pivot_table(time_genre, values = 'profit', index = ['genres_split'], columns = ['release_year'])
final_genre
```

```python
sns.set(rc = {'figure.figsize': (15,10)})
sns.heatmap(final_genre, cmap = 'YlGnBu', linewidths = .5)
plt.title('Genres by Profit per Year')
```
